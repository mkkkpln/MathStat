1) Построить модель линейной регрессии с указанными параметрами своими средствами (вместе со свободным коэффициентом).


Для построения модели линейной регрессии был использован датасет с данными о продолжительности, танцевальности и энергичности музыкальных треков, а также их популярности. Исследование проведено с уровнем значимости $\alpha = 0.05$.


Модель линейной регрессии имеет вид: $Y = \beta_0 + \beta_1 X_1 + \beta_2 X_2 + \beta_3 X_3 + \varepsilon,$ где:
- $Y$ - зависимая переменная (популярность),
- $\beta_0, \beta_1, \beta_2, \beta_3$ - коэффициенты регрессии,
- $X_1, X_2, X_3$ - предикторы (продолжительность, танцевальность, энергичность),
- $\varepsilon$ - случайная ошибка.



```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from scipy.stats import t, chi2

# Задаем уровень значимости
alpha = 0.05

# Датасет
df = pd.read_csv('song_data.csv')

# Построение модели линейной регрессии

# Матрица признаков X и вектор зависимой переменной Y
dur = np.array(df['song_duration_ms'])
danc = np.array(df['danceability'])
energy = np.array(df['energy'])
X = np.array([dur, danc, energy, np.ones(len(dur))]).T
Y = np.array(df['song_popularity'])
n = len(df)

df
```

Рассчитаем оценки наименьших квадратов для параметров.

1. Записываем модель линейной регрессии:  $y = X\beta+ϵ$

2. Мы хотим минимизировать сумму квадратов ошибок, которая записывается как:

  $S(\beta) = ∑{(Y_i-X_iс)^2} = (Y-X\beta)^T(Y-X\beta) = Y^TY-2\beta^TX^TY+\beta^TX^TX\beta$

3. Находим проихводную функции по $\beta$:

  $\frac{δS(\beta)}{δ\beta}=\frac{δ}{δс}(Y^TY-2\beta^TX^TY+\beta^TX^TX\beta)$

  $\frac{δS(\beta)}{δ\beta}=-2X^TY+2X^TX\beta$

4. Приравниваем производную к 0 для нахождения минимума

  $-2X^TY+2X^TX\beta = 0$

5. Получаем формулу оценки:

$\beta = (X^TX)^{-1}X^TY$

**Построение модели линейной регрессии своими средствами:**

Матрица: $A$ = $X^T$$X$,

Ее инверсия: $(X^TX)^{-1}$

Оценка наименьших квадратов для параметров:
$\beta$ = $(X^TX)^{-1}$$X^T$$Y$

```python
# Построение модели линейной регрессии своими средствами
A = np.dot(X.T, X)
invA = np.linalg.inv(A)
B = np.dot(np.dot(invA, X.T), Y)

print("Коэффициенты регрессии своими средствами:", B[0], B[1], B[2], B[3])
```

2) Рассчитать оценки наименьших квадратов для параметров и остаточной дисперсии.



```python
# Оценки наименьших квадратов для параметров и остаточной дисперсии
regression = LinearRegression().fit(X, Y)
print("С использованием готовой модели из sklearn:")
print("Коэффициенты регрессии (автоматические)", regression.coef_[0], regression.coef_[1], regression.coef_[2])
# Базовый уровень популярности трека, который не зависит от продолжительности, танцевальности и энергичности
print("Независимый уровень",regression.intercept_)
```

Теперь посчитаем оценку остаточной дисперсии, для этого есть формула:

${\beta}^2 = \cfrac{S^2({\beta})}{n - m}$


<!-- $S^2({\beta}) = (X{\beta} - Y)^T(X{\beta} - Y)$ -->

После этого:

$S^2({\beta}) = \frac{\sum{(X\beta_i - Y)^T (X\beta_i - Y)}}{n-m}$
   

В коде ниже:
- $X$ - это матрица признаков,
- $\beta$ - это вектор коэффициентов регрессии, полученных из модели.

Умножение $X @ C$ дает нам вектор прогнозных значений зависимой переменной (популярности трека), которые модель предсказывает на основе входных данных о продолжительности, танцевальности и энергичности трека.


- $Y$ - это истинные значения зависимой переменной (фактическая популярность трека). Вычитание $Y$ из прогнозных значений $X @ C$ дает нам вектор остатков, который представляет собой разницу между фактическими и предсказанными значениями.


- var: Это оценка остаточной дисперсии, которая является среднеквадратической ошибкой между фактическими и предсказанными значениями. Это мера разброса остатков относительно среднего.

```python
S = X @ B - Y
Ssq = np.sum(S**2)
m = X.shape[1] - 1
var = Ssq / (n - m)
print("Остаточная дисперсия:", var)
```

3. Вычислить коэффициент детерминации.

Коэффициент детерминации выглядит так:
$$R^2 =1 - \frac{\sigma^2}{\sigma^2_y}=1 - \cfrac{\sum\limits^n_{i = 1}(Y_i - \hat{Y})^2}{\sum\limits^n_{i = 1}(Y_i - \overline{Y})^2}$$
    
- Числитель - сумма квадратов отклонений фактических значений зависимой переменной ($Y_i$) от прогнозируемых значений ($\hat{Y}$) или условная дисперсия зависимой переменной


- Знаменатель - сумма квадратов отклонений фактических значений зависимой переменной ($Y_i$) от средней величины ($\overline{Y}$) или дисперсия случайной величины

Коэффициент показывает, какую долю общей вариации в данных объясняет модель.

```python
# Коэффициент детерминации
R = 0
#прогнозируемые значения
Yhat = X @ B


numerator = 0
for i in range(n):
    numerator += (Y[i] - Yhat[i]) ** 2

denominator = 0
for i in range(n):
    denominator += (Y[i] - np.mean(Y)) ** 2

R = 1 - numerator/denominator
print("Коэффициент детерминации:", R)
```

Коэффициент детерминации $R^2 = 0.01$ указывает на то,  что только около 1% вариации в данных объясняется моделью, что означает, что модель плохо соответствует данным.

**4. Построить доверительные интервалы для параметров модели и остаточной дисперсии.**


Доверительный интервал для параметров модели


$\hat{\beta}_i \pm t_{\alpha/2, \, n-3} \times SE(\hat{\beta}_i)$




Доверительного интервал для остаточной дисперсии:


$\frac{(n-3) \cdot s^2}{\chi^2_{\frac{\alpha}{2}}} < \sigma^2 < \frac{(n-3) \cdot s^2}{\chi^2_{1-\frac{\alpha}{2}}}$


```python
# Доверительные интервалы для параметров модели и остаточной дисперсии
leftb = []
rightb = []


def count_borders(i):
    leftb.append(B[i] + t.ppf(1 - alpha / 2, n - 3) * np.sqrt(np.linalg.inv(np.dot(X.T, X))[i][i]) * Ssq)
    rightb.append(B[i] - t.ppf(1 - alpha / 2, n - 3) * np.sqrt(np.linalg.inv(np.dot(X.T, X))[i][i]) * Ssq)

for i in range(3):
    count_borders(i)


print("Доверительные интервалы для параметров модели:")
print(leftb)
print(rightb)


# Рассчитываем квантили распределения хи-квадрат
alpha = 0.05  # уровень значимости
df = n - 3  # количество степеней свободы
chi2_lower = chi2.ppf(alpha / 2, df)
chi2_upper = chi2.ppf(1 - alpha / 2, df)
left_border = (n-3)*var/chi2_upper
right_border=(n-3)*var/chi2_lower

print("Доверительные интервалы для остаточной дисперсии:")
print(left_border)
print(right_border)
```
