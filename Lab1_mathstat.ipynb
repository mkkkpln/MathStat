# Задание 1

Экспоненциальное (или показательное) распределение — это абсолютно непрерывное распределение, моделирующее время между двумя последовательными свершениями одного и того же события.

Плотность вероятности экспоненциального распределения задается формулой:
\begin{equation}
f(x;\lambda) =
\begin{cases}
\lambda e^{-\lambda x}, & \text{если } x \geq 0 \\
0, & \text{если } x < 0
\end{cases}
\end{equation}
где $\lambda > 0$ - параметр интенсивности.

Функция распределения экспоненциального распределения имеет вид:
\begin{equation}
F(x;\lambda) =
\begin{cases}
1 - e^{-\lambda x}, & \text{если } x \geq 0 \\
0, & \text{если } x < 0
\end{cases}
\end{equation}



```python



import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy.stats import expon
#функция для расчета мат. ожидания, дисперсии и медианы выборки
def calculate_stats(sample):
    mean = np.mean(sample)
    variance = np.var(sample)
    median = np.median(sample)

    return mean, variance, median`

num_samples = 1000
num_columns = 1000
scale = 1
samples_matrix = np.random.exponential(scale, (num_samples,num_columns))
sample_means = np.mean(samples_matrix, axis=1) #средние значения по каждой строке матрицы
sample_variance = np.var(samples_matrix, axis=1)#дисперсия по каждой выборке
sample_quantiles = np.quantile(samples_matrix,0.5,axis=1)#квантили порядка 0.5

plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.hist(sample_means, bins=20, color='skyblue', edgecolor='black')
plt.axvline(np.mean(sample_means), color='r', linestyle='dashed', linewidth=1)
plt.title('Гистограмма средних значений')

plt.subplot(1, 3, 2)
plt.hist(sample_variance, bins=20, color='salmon', edgecolor='black')
plt.axvline(np.mean(sample_variance), color='r', linestyle='dashed', linewidth=1)
plt.title('Гистограмма выборочных дисперсий')

plt.subplot(1, 3, 3)
plt.hist(sample_quantiles, bins=20, color='lightgreen', edgecolor='black')
plt.axvline(np.mean(sample_quantiles), color='r', linestyle='dashed', linewidth=1)
plt.title('Гистограмма выборочных квантилей порядка 0.5')

plt.tight_layout()
plt.show()

sample_means_stats=calculate_stats(sample_means)
print(f"Для выборки мат.ожиданий: мат.ожидание: {sample_means_stats[0]} дисперсия: {sample_means_stats[1]} медиана: {sample_means_stats[2]}")

sample_variance_stats=calculate_stats(sample_variance)
print(f"Для выборки дисперсий: мат.ожидание: {sample_variance_stats[0]} дисперсия: {sample_variance_stats[1]} медиана: {sample_variance_stats[2]}")

sample_quantiles_stats=calculate_stats(sample_quantiles)
print(f"Для выборки квантилей: мат.ожидание: {sample_quantiles_stats[0]} дисперсия: {sample_quantiles_stats[1]} медиана: {sample_quantiles_stats[2]}")

#3ая часть задания
n = num_samples # размер выборки
statistic_1 = n * expon.cdf(np.sort(samples_matrix, axis=1)[:, 1]) #1 значение в выборке
statistic_2 = n * (1 - expon.cdf(np.sort(samples_matrix, axis=1)[:, -1]))#последнее значение в выборке


plt.figure(figsize=(12, 12))
plt.subplot(2, 2, 1)
plt.hist(statistic_1, bins=30, density=True, color='skyblue', alpha=0.7, label='Histogram')
x = np.linspace(0, 20, 100)
plt.legend()
plt.title('nF(X_(2))')


# Построим график плотности вероятности гамма-распределения с параметрами 2,1
plt.subplot(2, 2, 3)
plt.plot(x, stats.gamma.pdf(x, a=2, scale=1), 'r-', lw=2, label='Gamma(2, 1) PDF')

plt.subplot(2, 2, 2)
plt.hist(statistic_2, bins=30, density=True, color='red', alpha=0.7, label='Histogram')
x = np.linspace(0, 20, 100)
plt.legend()
plt.title('n(1 - F(X_(n)))')
plt.subplot(2, 2, 4)
# Построим график плотности вероятности экспоненциального распределения с коэф. 1
plt.plot(x, stats.expon.pdf(x, scale=1), 'b-', lw=2, label='Exponential(1) PDF')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Загрузка данных из файла
data = pd.read_csv('mobile_phones.csv')

# Количество моделей с двумя сим-картами
count_dual_sim = data[data['dual_sim'] == 1]['dual_sim'].count()
print(f"Количество моделей с 2мя сим-картами: {count_dual_sim}")
count_three_g = data[data['three_g'] == 1]['three_g'].count()
print(f"Количество моделей, поддерживающий 3-G: {count_three_g}")
max_cores = data['n_cores'].max()
print(f"Наибольшее число ядер у процессора : {max_cores}")

# Ёмкости аккумулятора
column_battery_power = data['battery_power']

#среднее
battery_power_mean = column_battery_power.mean()
print()
print(f"Выборочное среднее для всей совокупности ёмкости аккумулятора: {battery_power_mean}")

battery_power_variance = column_battery_power.var()
print(f"Выборочная дисперсия для всей совокупности ёмкости аккумулятора: {battery_power_variance}")

battery_power_median = column_battery_power.median()
print(f"Выборочная медиана для всей совокупности ёмкости аккумулятора: {battery_power_median}")

battery_power_quantile = column_battery_power.quantile(2/5)
print(f"Выборочная квантиль порядка 2.5 для всей совокупности ёмкости аккумулятора: {battery_power_quantile}")

# Построение графиков

plt.figure(figsize=(6, 6))
# Эмпирическая функция распределения
sns.ecdfplot(data=column_battery_power, label='Empirical CDF')
plt.title('Эмпирическая функция распределения')
plt.show()
plt.figure(figsize=(6, 6))

# Гистограмма general
sns.histplot(data=column_battery_power, bins=10, kde=False,stat='density', color='pink')
plt.title('Гистограмма ёмкости аккумлятора')
plt.xlabel('Ёмкость аккумулятора')
plt.ylabel('Вероятность')

plt.show()

# Ёмкости аккумулятора моделей, поддерживающих wifi
battery_power_with_wifi = data[data['wifi'] == 1]['battery_power']
# Ёмкости аккумулятора моделей, не поддерживаюющих wifi
battery_power_wo_wifi = data[data['wifi'] == 0]['battery_power']
battery_power_with_wifi_mean = battery_power_with_wifi.mean()
print(f"Выборочное среднее для аккумуляторов, поддерживающих wifi: {battery_power_with_wifi_mean}")

battery_power_with_wifi_variance = battery_power_with_wifi.var()
print(f"Выборочная дисперсия для аккумуляторов, поддерживающих wifi: {battery_power_with_wifi_variance}")

battery_power_with_wifi_median = battery_power_with_wifi.median()
print(f"Выборочная медиана для аккумуляторов, поддерживающих wifi: {battery_power_with_wifi_median}")

battery_power_with_wifi_quantile = battery_power_with_wifi.quantile(2/5)
print(f"Выборочная квантиль порядка 2.5 для аккумуляторов, поддерживающих wifi: {battery_power_with_wifi_quantile}")
plt.figure(figsize=(8, 8))
# Гистограмма with wifi
sns.histplot(data=battery_power_with_wifi, bins=10, kde=False, color='pink')
plt.title('Гистограмма ёмкости аккумлятора с wifi')
plt.xlabel('Ёмкость аккумулятора')
plt.ylabel('Количество')

plt.show()

battery_power_wo_wifi_mean = battery_power_wo_wifi.mean()
print(f"Выборочное среднее для аккумуляторов, поддерживающих wifi: {battery_power_wo_wifi_mean}")

battery_power_wo_wifi_variance = battery_power_wo_wifi.var()
print(f"Выборочная дисперсия для аккумуляторов, поддерживающих wifi: {battery_power_wo_wifi_variance}")

battery_power_wo_wifi_median = battery_power_wo_wifi.median()
print(f"Выборочная медиана для аккумуляторов, поддерживающих wifi: {battery_power_wo_wifi_median}")

battery_power_wo_wifi_quantile = battery_power_wo_wifi.quantile(2/5)
print(f"Выборочная квантиль порядка 2.5 для аккумуляторов, поддерживающих wifi: {battery_power_wo_wifi_quantile}")



plt.figure(figsize=(6, 6))
# Гистограмма without wifi
sns.histplot(data=battery_power_wo_wifi, bins=10, kde=False, color='pink')
plt.title('Гистограмма ёмкости аккумлятора без wifi')
plt.xlabel('Ёмкость аккумулятора')
plt.ylabel('Количество')

plt.show()
plt.figure(figsize=(6, 6))
#general box-plot

plt.boxplot(data['battery_power'], vert=False)
plt.xlabel('Ёмкость аккумулятора')
plt.title('Box Plot для ёмкости аккумулятора')
plt.show()

#box-plot №2

plt.figure(figsize=(6, 6))
sns.boxplot(x='wifi', y='battery_power', showfliers=True,showcaps=True, data=data)

plt.title('Box Plot для ёмкости аккумулятора с поддержкой Wi-Fi ')
plt.xlabel('Поддержка Wi-Fi ')
plt.ylabel('Ёмкость аккумулятора')




plt.tight_layout()
plt.show()

```
